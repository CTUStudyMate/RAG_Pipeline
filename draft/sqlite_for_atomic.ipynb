```python
import sqlite3

conn = sqlite3.connect("rag_storage93_01.db")
cursordb = conn.cursor()
conn.commit()
```

```python
cursordb.execute("""
CREATE TABLE IF NOT EXISTS documents (
    id TEXT PRIMARY KEY,
    title TEXT,
    source_path TEXT,
    created_at DATETIME DEFAULT CURRENT_TIMESTAMP
);
""")

cursordb.execute("""
CREATE TABLE IF NOT EXISTS atomic_elements (
    id TEXT PRIMARY KEY,
    document_id TEXT NOT NULL,

    type TEXT NOT NULL,              -- picture / section_header / page_footer / paragraph
    content TEXT,                    -- text hoặc base64 image
    token_count INTEGER,

    atomic_order INTEGER NOT NULL,   -- thứ tự trong tài liệu

    -- metadata phổ biến
    description TEXT,
    level INTEGER,                   -- cho heading
    heading_type TEXT,               -- main / sub
    path TEXT,

    created_at DATETIME DEFAULT CURRENT_TIMESTAMP,

    FOREIGN KEY (document_id) REFERENCES documents(id)
);
""")

cursordb.execute("""
CREATE INDEX IF NOT EXISTS idx_atomic_doc_order
ON atomic_elements(document_id, atomic_order);
""")

conn.commit()
```

```python
cursordb.execute("SELECT * FROM atomic_elements;")
rows = cursordb.fetchall()

for row in rows:
    print(row)
```
